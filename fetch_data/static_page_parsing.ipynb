# Static HTML

```python
def parse_static_HTML(url, headers=None, **kwargs):
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    return soup
```

# Session HTML

```python
def parse_session_HTML(url, headers=None, **kwargs):
    session = HTMLSession()
    response = session.get(url, headers=headers, timeout=7)
    soup = BeautifulSoup(response.content, 'html.parser')
    return soup
```

# Dynamic Website

```python
def parsing_by_dynamic(url, headers=None):
    
 
    options = webdriver.EdgeOptions()
    options.use_chromium = True  
    options.add_argument('--headless') 
    driver = webdriver.Edge(options=options) # if the error of Edge version, download the newest 
    #driver = webdriver.Chrome()
    driver.get(url)
    driver.implicitly_wait(5)  # waiting seconds
    html = driver.page_source
    driver.quit()
    soup = BeautifulSoup(html, 'html.parser')
    #soup.find_all("span", class_='job-tile__title')
    
    return soup
```
